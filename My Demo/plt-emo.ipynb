# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importing fonts

import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph (dpi specifies pixels, 300 is recommended)

plt.figure(figsize=(5,6), dpi=150)

# Line 1

# Keyword Argument Notation
# plt.plot(x,y, label='2x',  color='red', linewidth=2, marker='.', markersize=15, markeredgecolor='red', linestyle='--')

# Shorthand Notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'bs-', label='2x')

## Line 2

# Select interval
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:],x2[5:]**2, 'r--')

# Add a title (specify font paramenters with fontdict)
plt.title('Our First Graph!', fontdict={ 'fontname': 'Mukta', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname':'Arial', 'fontsize': 10})
plt.ylabel('Y Axis')

# x,y axis tickmarks (scale the graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# add a legend
plt.legend()

# save figure (dpi >300 for high resolution)
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')


plt.figure(figsize=(6,4))

plt.show()
```
